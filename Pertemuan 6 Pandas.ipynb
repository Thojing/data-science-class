## Missing Value

### Impute/Mengatasi data yang kosong
1. Delete data
2. Isi data dengan mean, median, modus
3. Isi data dengan data sembarangan

```python
df.isna().sum()
```

```python
#Menghapus data kosong
df.dropna()
```

```python
#Mengisi data kosong dengan rata-rata (mean)
df.fillna(df.mean()).head()
```

```python
#Mengisi data dengan nilai sembarangan
df.fillna('310.0').head()
```

```python
#Mengisi data sembarangan lebih dari satu variabel
isi = {'carbo':10.0,'sugars':12.0,'potass':320.0}
df.fillna(isi).head()
```

```python
df.fillna(isi,inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.to_csv('Data/CerealEdit.csv')
```
