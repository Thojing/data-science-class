## DATA AUDIBLE

```python
#Import Data
audible = pd.read_csv('Data/TO/audible.csv')
audible.head()
```

```python
#Melihat Info Data
audible.info()
```

```python
#Melihat Jumlah Data Hilang
audible.isna().sum()
```

```python
#Membuang Kolom Tidak Penting
audible.drop(columns= ['time'],inplace = True)
audible.head()
```

```python
new_names = {'name' : 'Name',
            'author' : 'Author',
            'narrator' : 'Narrator',
            'stars' : 'Stars',
            'releasedate' : 'Release',
            'language' : 'Language',
            'price' : 'Price'}
```

```python
#Mengganti Nama Kolom
audible.rename(columns= new_names, inplace = True)
audible.head()
```

```python
#Merapikan Data dengan str
audible['Author'] = audible.Author.str.replace('Writtenby:', ' ')
audible.head()
```

```python
#Merapikan Data dengan str
audible['Narrator'] = audible.Narrator.str.replace('Narratedby:', ' ')
audible.head()
```

```python
#Merapikan Fields Data
audible['Release'] = pd.to_datetime(audible['Release'])
audible.head()
```

```python
#Merapikan Fields Data
regex = r'^(\d{1})'
extr = audible['Stars'].str.extract(r'^(\d{1})', expand = False)
extr.head()
```

```python
#Merapikan Fields Data
audible['Stars'] = pd.to_numeric(extr)
audible['Stars'].dtype
```
