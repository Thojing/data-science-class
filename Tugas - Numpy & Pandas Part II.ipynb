## Missing Value

### Impute/Mengatasi Data yang Kosong

1. Menghapus Data
2. Mengisi data dengan mean, median, atau modus
3. Mengisi data dengan data acak

```python
#Import Packages
import numpy as np
import pandas as pd
```

```python
#Import Data
tugas = pd.read_csv('Data/IBL_2022.csv')
tugas
```

```python
tugas.isna().sum()
```

```python
#Menghapus Data Kosong
tugas.dropna()
```

```python
#Mengisi data kosong dengan rata-rata (mean)
tugas.fillna(tugas.mean()).head()
```

```python
#Mengisi data sembarangan lebih dari satu variabel
isi = {'Points':50.0,'Assist':12.0,'Rebound':20.0,'Game':18.0}
tugas.fillna(isi).head()
```

```python
tugas.fillna(isi,inplace=True)
tugas
```

```python
tugas.isna().sum()
```

```python
tugas.to_csv('Data/IBL_2022Edit.csv')
```

## GroupBy

```python
#Membuka Dataset
tugas2 = pd.read_csv ('Data/IBL_2022Edit.csv',index_col= 'Unnamed: 0')
tugas2
```

```python
tugas2.columns
```

```python
tugas2.info()
```

```python
#Membuat Groupby 
tugas2.groupby('Tim').mean(numeric_only=True)
```

```python
tugas2.groupby('Position').mean(numeric_only=True)
```

```python
tugas2.groupby(['Tim','Position']).mean(numeric_only=True)
```

```python
tugas2.groupby(['Tim','Position']).count()
```
