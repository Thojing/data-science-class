```python
%matplotlib inline
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import os
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import sklearn.metrics
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
```

```python
#Import Data
df = pd.read_csv('Data/tree_addhealth.csv')
df.head(10)
```

```python
#Cek Missing Value
df.isnull().sum()
```

```python
#Drop Missing Value
df_baru = df.dropna()
```

```python
#Describe data
df_baru.describe()
```

```python
#Cek Kembali Missing Value
df_baru.isnull().sum()
```

```python
#Cek Data Numerik
num_of_numerical_cols = df_baru._get_numeric_data().columns.shape[0]
print(num_of_numerical_cols, 'numerical columns')
name_of_numerical_cols = df_baru._get_numeric_data().columns
print(name_of_numerical_cols)
```

```python
#Split Data
x = df_baru[['BIO_SEX', 'HISPANIC', 'WHITE', 'BLACK', 'NAMERICAN', 'ASIAN', 'age', 'ALCEVR1', 'ALCPROBS1', 'marever1', 'cocever1', 'inhever1',
       'cigavail', 'DEP1', 'ESTEEM1', 'VIOL1', 'PASSIST', 'DEVIANT1',
       'SCHCONN1', 'GPA1', 'EXPEL1', 'FAMCONCT', 'PARACTV', 'PARPRES']]
y = df_baru.TREG1

x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=.4, random_state= 42)
```

```python
print('x_train = ', x_train.shape)
print('x_test = ', x_test.shape)
print('y_train = ', y_train.shape)
print('y_test = ', x_test.shape)
```

```python
#Build Model
classifier = RandomForestClassifier(n_estimators=20, random_state = 0)
classifier.fit(x_train, y_train)

predictions = classifier.predict(x_test)
```

```python
#confusion matrix
confusion_array = sklearn.metrics.confusion_matrix(y_test,predictions)
print(confusion_array)
```

```python
print('TP = ', confusion_array[0,0]) #True Positive
print('FP = ', confusion_array[0,1]) #False Positive
print('FN = ', confusion_array[1,0]) #False Negative
print('TN = ', confusion_array[1,1]) #True Negative
```

```python
#Akurasi
classifier.score(x_test, y_test)
```

```python
sklearn.metrics.accuracy_score(y_test, predictions)
```
