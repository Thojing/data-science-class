```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
#Import data
df = pd.read_csv('Data/melb_data.csv', sep= ';')
df.head()
```

```python
#Info data
df.info()
```

```python
#Mengambil 1500 baris pertama
df = df[:1500]
df.head()
```

```python
#Info data
df.info()
```

```python
#Jumlah data kosong
df.isna().sum()
```

```python
#Tipe data
df.dtypes.value_counts()
```

```python
#Merapikan baris Date
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
regex = r'^(\d{4})'
```

```python
#Merapikan baris Postcode
extr = df['Postcode'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['Postcode'] = pd.to_numeric(extr)
df['Postcode'].dtype
```

```python
df.head()
```

```python
#Merapikan baris Price
df['Price'] = df['Price'].astype(int)
df.head()
```

```python
#Merapikan baris Bathroo
df['Bathroom'] = df['Bathroom'].astype(int)
df.head()
```

```python
#Merapikan baris Bedroom2
df['Bedroom2'] = df['Bedroom2'].astype(int)
df.head()
```

```python
#Merapikan baris Car
df['Car'] = df['Car'].astype(int)
df.head()
```

```python
#Merapikan baris YearBuilt
extr = df['YearBuilt'].astype(str).str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['YearBuilt'] = pd.to_numeric(extr)
df['YearBuilt'].dtype
```

```python
df.head()
```

```python
#Merapikan baris Propertycount
extr = df['Propertycount'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['Propertycount'] = pd.to_numeric(extr)
df['Propertycount'].dtype
```

```python
df.head()
```

```python
#Menambahkan data yang kosong pada variabel BuildingArea & YearBuilt dengan Frequent Category
imp = SimpleImputer(strategy='most_frequent')
```

```python
df = imp.fit_transform(df)
```

```python
print(df)
```
