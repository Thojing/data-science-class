## Pre-Modeling

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Label Encoder
LEncoder = LabelEncoder()

df['Wilayah'] = LEncoder.fit_transform(df['Wilayah'])
```

```python
df.info()
```

```python
df['Wilayah']
```

```python
X = df.drop('Wilayah',axis=1).values
y = df['Wilayah'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.4, random_state=42)
```

```python
sc = StandardScaler()

X_train = sc.fit_transform(x_train)
X_test= sc.transform(x_test)
```

## Modeling

```python
# Modeling

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
```

```python
# Random Forest Classifier

rf = RandomForestClassifier()

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
```

```python
# Decision Tree Classifier

dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
```

```python
# XGB Classifier

XGB = XGBClassifier() 

XGB.fit(X_train, y_train)

y_pred_XGB = XGB.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score, recall_score, confusion_matrix, precision_recall_fscore_support
from sklearn.metrics import roc_auc_score,auc,f1_score
from sklearn.metrics import precision_recall_curve,roc_curve
import sklearn.metrics as metrics
```

```python
models = []

Z = [DecisionTreeClassifier(),XGBClassifier(), RandomForestClassifier()]


X = ["DecisionTreeClassifier", "RandomForestClassifier" ,"XGB"]

for i in range(0,len(Z)):
    model = Z[i]
    model.fit( X_train , y_train )
    pred = model.predict(X_test)
    models.append(accuracy_score(pred , y_test))
```

```python
d = { "Accuracy" : models , "Algorithm" : X }
data_frame = pd.DataFrame(d)
data_frame
```
