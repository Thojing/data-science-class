```python
plt.plot(x,y);
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu Y')
plt.title('Data Penyakit');
```

```python
plt.subplot(121)
plt.plot(x,y, color = 'blue')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu Y')

plt.subplot(122)
plt.plot(y,x, color = 'red')
plt.xlabel('Sumbu x2')
plt.ylabel('Sumbu y2')
plt.tight_layout()
plt.show()
```
